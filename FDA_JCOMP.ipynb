## DATA CLEANING
1. remove retweet indicator "!!!RT"/ "RT" 
2. CONVERT &amp to "and" if left and right of it are
 {letters, ","} 
3. remove usernames "@ stuff", after removing if tweet length zero then remove the row 
4. **remove the numbers starting with &# (emojis)
5. remove links 
6. remove the hashtag symobl, but preserve the text ##
7. &#8220 ->(OPENING QUOTATION) &#8221 CLOSING QUOTATION
8. REMOVE DUPLICATE ROWS
9. remove asterisks 
10. remove :) uwu ;) 
11. remove "......" 
12. DIS->this Dat -> that  
13. &lt;	&gt; &le; &ge; remove 
14. cuz ->  because 
15. 'fuc' or 'fuk' or 'fucc' -> fuck; fuccn-> fucking 
16. DA-> the 
17. ya -> you 
18. hahaha or haha or hahahahahah or hahaha or hahahaa-> haha 
19. yall or y'all -> you all 

```python
a=data$tweet
```

```python
for(i in 1:length(a)){
        a[i]=str_remove_all(a[i],"(!)*") #removing retweet indication
        a[i]=str_remove_all(a[i],"(RT)") #removing retweet indication
        a[i]=str_remove_all(a[i],"@([A-Za-z0-9_:]+)") #removing user names
        a[i]=str_remove_all(a[i],"http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+#]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+") #removing user links
        a[i]=str_remove_all(a[i],"(&#)[0-9]*(;)") #removing html ascii
        a[i]=str_remove_all(a[i],"#")#removing hashtags but preserving the texts
        a[i]=str_replace_all(a[i],'( cuz )|( CUZ )|(cuz )|( CUZ )', ' because ')
        a[i]=str_replace_all(a[i],'(&amp)', ' and ')
        a[i]=str_replace_all(a[i],'( dis )|( Dis )|( DIS )|(dis )|(Dis )|(DIS )', ' this ')
        a[i]=str_replace_all(a[i],'( dat )|( Dat )|( DAT )|(dat )|(Dat )|(DAT )', ' that ')
        a[i]=str_replace_all(a[i],'( da )|( Da )|( DA )|(da )|(Da )|(DA )', ' the ')
        a[i]=str_replace_all(a[i],'( ya )|( Ya )|( YA )|(ya )|(Ya )|(YA )', ' you ')
        a[i]=str_replace_all(a[i],'( yall )|( y\'all )|(yall )|(y\'all )|(ya\'ll)', ' you all ')
        a[i]=str_replace_all(a[i],'( fuc )|( fuk )|( fucc )|( Fuc )|( Fuk )|( Fucc )|(fuc )|(fuk )|(Fuc )|(fucc )|(Fuk )|(Fucc )|(fuc )', ' fuck ')
        a[i]=str_replace_all(a[i],'( fuccn )|( Fuccn )', ' fucking ')
        a[i]=str_remove_all(a[i],'(ha){2,}')
        a[i]=str_remove_all(a[i],'(\\*){2,}')
        a[i]=str_replace_all(a[i],'(\'re )', ' are ')
        a[i]=str_replace_all(a[i],'(\'ve )', ' have ')
        a[i]=str_replace_all(a[i],'(\'d )', ' had ')
        a[i]=str_replace_all(a[i],'(\'s )', ' is ')
        a[i]=str_remove_all(a[i],"(\\.){2,}") #removing periods
        a[i]=str_remove_all(a[i],"(&lt;)|(&gt;)|(&le;)|(&ge;)")
        a[i]=str_replace_all(a[i],'(\'ll )', ' will ')
        a[i]=str_replace_all(a[i],'(\'m )', ' am ')
        a[i]=str_replace_all(a[i],'(can\'t )', 'cannot ')
        
        a[i]=str_replace_all(a[i],'(n\'t )', ' not ')
        
        
      }

```

```python
tfv <- TfIdfVectorizer$new(min_df = 0.01,remove_stopwords = TRUE, ngram_range = c(1,1))
data.tfv <- tfv$fit_transform(data2$tweet)
```

## SVM + TF IDF

### binary

#### TRAIN TEST SPLIT: 75:25

```python
data3= data.frame(cbind(data.tfv, target = data2$bin))
data3$target = factor(data3$target, levels = c(0, 1))
set.seed(123)
split = sample.split(data3$target, SplitRatio = 0.75)
training_set = subset(data3, split == TRUE)
test_set = subset(data3, split == FALSE)
```

```python
classifier = svm(formula = target ~ .,
                 data = training_set,
                 type = 'C-classification',
                 kernel = 'linear')
```

#### prediction for binary

```python
y_pred = predict(classifier, newdata = test_set[-96])
```

```python
cm=confusionMatrix(y_pred, test_set$target)
```

#### CREATING DOCUMENT TERM MATRIX FOR HASH VECTORIZER

```python
get_matrix <- function(text) {
  it <- itoken(text, progressbar = FALSE)
  create_dtm(it, vectorizer = hash_vectorizer())
}
```

```python
dtm_train <- get_matrix(training_set_bin_lime$tweet)
str(dtm_train)
```

```python
dtm_test <- get_matrix(test_set_bin_lime$tweet)
str(dtm_test)
```

```python
svm <- svm(x = dtm_train,y = training_set_bin_lime$target,
                 type = 'C-classification',
                 kernel = 'linear')
```

#### PREDICTION,CONFUSION MATRIX,METRICS

```python
pred <- predict(svm, dtm_test)

confusionMatrix(test_set_bin_lime$target,
                pred)
```

# Xgboost

## Binary

```python
data2$target = factor(data2$bin, levels = c(0, 1))
set.seed(123)
split_bin_lime = sample.split(data2$bin, SplitRatio = 0.75)
training_set_bin_lime = subset(data2, split_bin_lime == TRUE)
test_set_bin_lime = subset(data2, split_bin_lime == FALSE)
```

```python
model_type.xgb.Booster <- function(x){
  return("classification")
}
```

### DOCUMENT TERM MATRIX

```python
get_matrix <- function(text) {
  it <- itoken(text, progressbar = FALSE)
  create_dtm(it, vectorizer = hash_vectorizer())
}
```

```python
dtm_train <- get_matrix(training_set_bin_lime$tweet)
str(dtm_train)
```

```python
dtm_test <- get_matrix(test_set_bin_lime$tweet)
str(dtm_test)
```

### MODEL TRAINING 

```python
xgb_model <- xgb.train(params = list(max_depth = 10, 
                                      eta = 0.2, 
                                      objective = "binary:logistic",
                                      eval_metric = "error", nthread = 1),
                        data = xgb.DMatrix(dtm_train, 
                                           label =as.numeric(as.character(training_set_bin_lime$target))),
                        nrounds = 500)
```

### PREDICTION

```python
pred <- predict(xgb_model, dtm_test)
temp=pred
temp[temp>=0.5]=1
temp[temp<0.5]=0
```

```python
confusionMatrix(test_set_bin_lime$target,
                factor(temp,levels = c(0,1)))
```

## Multiclass

```python
data5=data2
data5$target = factor(data5$class, levels = c(0, 1,2))
set.seed(123)
split_mul_lime = sample.split(data5$class, SplitRatio = 0.75)
training_set_mul_lime = subset(data5, split_mul_lime == TRUE)
test_set_mul_lime = subset(data5, split_mul_lime == FALSE)
```

```python
dtm_train1 <- get_matrix(training_set_mul_lime$tweet)
str(dtm_train1)
```

```python
dtm_test1 <- get_matrix(test_set_mul_lime$tweet)
str(dtm_test1)
```

```python
numberOfClasses <- length(unique(training_set_mul_lime$target))
xgb_model1 <- xgb.train(params = list(max_depth = 10, 
                                      eta = 0.2, 
                                      objective = "multi:softprob",
                                      eval_metric = "error", nthread = 1,"num_class" = numberOfClasses),
                        data = xgb.DMatrix(dtm_train1, 
                                           label =as.numeric(as.character(training_set_mul_lime$target))),
                        nrounds = 500)
```

```python
pred1 <- predict(xgb_model1, dtm_test1,reshape=T)
mul_pred <- data.frame(pred1) %>%
  mutate(max_prob = (max.col(., ties.method = "last")-1))
```

```python
confusionMatrix(test_set_mul_lime$target,
                factor(mul_pred$max_prob,levels = c(0,1,2)))
```
